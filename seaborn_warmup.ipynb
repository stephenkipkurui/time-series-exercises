````python
# Horsepower by origin 
# Ci = None removes the boothstrapping confidence interval and error bars. 

ax = sns.barplot(data = df, x = 'origin', y = 'horsepower', ci = None)
````

````python
# Scatter plot 
# Hue = weight - add index of weighted colors
# paletter = Greens - changes the color

sns.scatterplot(data = df, x = 'horsepower', y = 'displacement', hue = 'weight', palette='Greens')
plt.show()
````

````python
# Scatter plot 
# Hue = weight - add index of weighted colors
# paletter = Greens - changes the color
# Hue = cylinders- change color by cylinders
sns.scatterplot(data = df, x = 'mpg', y = 'displacement', palette='tab10', hue = 'cylinders')
None
````
